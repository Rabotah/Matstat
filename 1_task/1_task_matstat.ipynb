```python
import numpy as np
import matplotlib.pyplot as plt
```

### выборка

```python
N = 100
teta = 142
np.random.seed(44)
X = np.random.randint(0, teta, N)
X                                    
```

### оценки

```python
t1 = [2*np.sum(X[:n])/n for n in range(1,N+1)]
t2 = [(n+1)*np.min(X[:n]) for n in range(1,N+1)]
t3 = [(n+1)*np.max(X[:n])/n for n in range(1,N+1)]
t4 = [(np.min(X[:n]) + np.max(X[:n])) for n in range(1,N+1)]
t5 = [(X[0] + 1/(n-1)*np.sum(X[1:n])) for n in range(2,N+1)]
t5.insert(0, X[0])
```

### графики

```python
fig, ax = plt.subplots(figsize = (15, 5))
x = np.arange(1, N+1, dtype = int)
y1 = [abs(s - teta) for s in t1]
y2 = [abs(s - teta) for s in t2]
y3 = [abs(s - teta) for s in t3]
y4 = [abs(s - teta) for s in t4]
y5 = [abs(s - teta) for s in t5]
plt.grid()
ax.plot(x, y1, 'violet', label = '\N{greek small letter theta}1')
ax.plot(x, y2, 'blue', label = '\N{greek small letter theta}2')
ax.plot(x, y3, 'green', label = '\N{greek small letter theta}3')
ax.plot(x, y4, 'orange', label = '\N{greek small letter theta}4')
ax.plot(x, y5, 'red', label = '\N{greek small letter theta}5')
ax.legend()
plt.show()
```
